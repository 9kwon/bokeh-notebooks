## Histograms

```python
import bokeh
bokeh.load_notebook()
```

```python
from collections import OrderedDict

import numpy as np
from bokeh.charts import Histogram

mu, sigma = 0, 0.5 # NOTE: you can tinker with these values if you like

# sample the distribution
normal = np.random.normal(mu, sigma, 1000)

# Histogram can take a dictionary as input
normal_dist = OrderedDict(normal=normal)

hist = Histogram(normal_dist, bins=50, mu=mu, sigma=sigma,
                 title="kwargs, dict_input", ylabel="frequency", legend="top_left",
                 width=400, height=350, notebook=True)
hist.show()
```

```python
import pandas as pd
df = pd.DataFrame(normal_dist)

# Histogram can also take a pandas DataFrame as input
hist = Histogram(df, bins=50, mu=mu, sigma=sigma,
                 title="no_tools, df_input", ylabel="frequency", legend="top_left",
                 tools=True, width=400, height=350, notebook=True)
hist.show()
```

```python
# We can also plot several distributions in the same figure with a dictionary or a DataFrame
mu, sigma = 0, 0.5

normal = np.random.normal(mu, sigma, 1000)
lognormal = np.random.lognormal(mu, sigma, 1000)

distributions = OrderedDict(normal=normal, lognormal=lognormal)

hist = Histogram(distributions, bins=50, notebook=True)
hist.title("chained_methods, dict_input").ylabel("frequency").legend(True).width(400).height(350).show()
```

```python
import pandas as pd
df = pd.DataFrame(distributions)

hist = Histogram(df, bins=50, notebook=True)
hist.title("chained_methods, df_input").ylabel("frequency").width(400).height(350).show()
```

```python
# EXERCISE: Add new plots for other distributions. Some code
# has be included for the Gamma, Beta, and Weibull distributions.

#Code for Gamma Distribution
k, theta = 1.0, 2.0
gamma_sample = np.random.gamma(k, theta, 1000)

#Code for Beta Distribution
alpha, beta = 2.0, 2.0
beta_sample = np.random.beta(alpha, beta, 1000)

#Code for Weibull Distribution
lam, k = 1, 1.25
weisbull_sample = lam*(-np.log(np.random.uniform(0, 1, 1000)))**(1/k)
```

```python
#Note: You can either use an OrderedDict or a pandas DataFrame
gamma_dist = OrderedDict(gamma_sample=gamma_sample)
beta_dist = OrderedDict(beta_sample=beta_sample)
weisbull_dist = OrderedDict(weisbull_sample=weisbull_sample)

#import scipy

#Gamma Histogram
gamma_hist = Histogram(gamma_dist, bins=50, notebook=True)
# compute ideal values
#x = np.linspace(0, 20.0, 1000)
#pdf = x**(k-1) * np.exp(-x/theta) / (theta**k * scipy.special.gamma(k))
#cdf = scipy.special.gammainc(k, x/theta) / scipy.special.gamma(k)

gamma_hist.title("Gamma Distribution (k=1, θ=2)").ylabel("frequency").legend(True).width(400).height(350).show()

#line(x, pdf, line_color="#D95B43", line_width=8, alpha=0.7, legend="PDF")
#line(x, cdf, line_color="white", line_width=2, alpha=0.7, legend="CDF")

#Beta Histrogram
beta_hist = Histogram(beta_dist, bins=50, notebook=True)
beta_hist.title("Beta Distribution (α=2, β=2)").ylabel("frequency").legend(True).width(400).height(350).show()

#Weisbull Histrogram
weisbull_hist = Histogram(weisbull_dist, bins=50, notebook=True)
weisbull_hist.title("Weibull Distribution (λ=1, k=1.25)").ylabel("frequency").legend(True).width(400).height(350).show()
```
